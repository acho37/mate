```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tramo = "tramo"+input("qu√© tramo: ")
print(tramo)
archivo_excel = pd.read_excel(open('1984_2.xlsx', 'rb'),
              sheet_name= tramo) 
values = archivo_excel['puntos'].values
print(values)
xy = ['X', 'Y']
X = ['X']
Y = ['Y']
columnas_seleccionadas = archivo_excel[xy]
cx = archivo_excel[X]
cy = archivo_excel[Y]
print(columnas_seleccionadas)
qp = len(cy)
xcx = np.array(cx)
ycy = np.array(cy)
print("cantidad de puntos (qp): ",qp)
```

```python
A = np.zeros((qp,qp))
for i in range(qp):
    A[:,i] = cx['X']**i
y = cy['Y']
```

```python
coef = np.linalg.solve(A,y)
coef
```

```python
plt.scatter(cx,cy)
x = np.linspace(np.min(cx),np.max(cx),100)
y = x*0
for i in range(qp):
    y += coef[i]*x**i
plt.plot(x,y)
```

```python
matrizx = []
nrow = 0
while nrow < qp:
    exp = 0
    while exp < qp:
        valor_row = (pow(xcx[nrow],exp))
        matrizx.extend(valor_row)
        exp += 1
    nrow += 1
if exp >= qp: 
    exp -= 1
np.array(matrizx)
print(matrizx)

```
